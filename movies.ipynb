```python
import pandas as pd
```

```python
df=pd.read_csv("movie.csv")
```

```python
df.head()
```

```python

movies_df = df.assign(genres=df['genres'].str.split('|')).explode('genres')
movies_df.head()
```

```python
genre_movies = {}
for genre in movies_df['genres'].unique():
    genre_movies[genre] = movies_df[movies_df['genres'] == genre]['title'].tolist()

print(genre_movies.get("Action"))
```
